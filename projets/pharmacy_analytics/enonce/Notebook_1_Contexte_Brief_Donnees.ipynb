## 🔍 Contexte opérationnel

L'hôpital gère **10 services** (Urgences, Cardiologie, Pédiatrie, Chirurgie, etc.) et **25 médicaments critiques**.

La pharmacie centrale reçoit des commandes de **5 fournisseurs** dont les délais varient de 3 à 21 jours. Elle doit anticiper la demande sur **4 semaines** pour passer les commandes à temps.

**Le problème est double :**
- **Ruptures de stock** → risque clinique direct pour les patients
- **Surstockage** → gaspillage financier + péremption des médicaments

## ❓ Les 5 questions business à résoudre

En tant que Data Analyst, ta mission est de répondre à ces questions :

1. **Analyse historique** — Quels médicaments et quels services consomment le plus ? Y a-t-il des tendances et des saisonnalités ?
2. **Diagnostic des ruptures** — Quand, pourquoi et sur quels médicaments les ruptures se produisent-elles ?
3. **Performance fournisseurs** — Quels fournisseurs sont les plus fiables en termes de délai et de service ?
4. **Prévision** — Peut-on prédire la consommation des 4 prochaines semaines par médicament ?
5. **Recommandations** — Quel niveau de stock de sécurité recommander pour chaque médicament critique ?

## 📁 Présentation des données

Tu disposes de **6 fichiers CSV** couvrant la période **Janvier 2022 → Juin 2024** (30 mois).

### 📄 Fichier 1 — `medicaments.csv` (25 lignes)
Référentiel de tous les médicaments gérés par la pharmacie.

| Colonne | Type | Description |
|---|---|---|
| `id_medicament` | string | Identifiant unique (ex: MED001) |
| `nom` | string | Dénomination commune internationale |
| `categorie` | string | Classe thérapeutique |
| `forme` | string | Forme galénique (Comprimé, Injectable, Spray...) |
| `stock_actuel` | int | Stock disponible en unités |
| `prix_unitaire_euro` | float | Prix d'achat unitaire en euros |
| `stock_securite` | int | Seuil minimum à ne jamais franchir |
| `stock_max` | int | Seuil maximum de stockage |
| `medicament_critique` | bool | True si rupture = risque clinique élevé |

### 📄 Fichier 2 — `consommations.csv` (~53 000 lignes)
Le cœur du projet. Chaque ligne = une consommation journalière d'un médicament dans un service.

| Colonne | Type | Description |
|---|---|---|
| `id_consommation` | string | Identifiant unique |
| `date` | date | Date de consommation (YYYY-MM-DD) |
| `id_medicament` | FK | Référence vers medicaments.csv |
| `id_service` | FK | Référence vers services.csv |
| `quantite_consommee` | int | Nombre d'unités consommées ce jour |
| `cout_euro` | float | Coût total de la consommation |

### 📄 Fichier 3 — `commandes_fournisseurs.csv` (~964 lignes)
Toutes les commandes passées aux fournisseurs.

| Colonne | Type | Description |
|---|---|---|
| `id_commande` | string | Identifiant unique |
| `date_commande` | date | Date de passage de commande |
| `id_medicament` | FK | Médicament commandé |
| `id_fournisseur` | FK | Fournisseur choisi |
| `quantite_commandee` | int | Quantité commandée |
| `montant_total_euro` | float | Montant total de la commande |
| `statut` | string | Livrée / En cours / Annulée |
| `date_livraison_reelle` | date | Date effective de livraison |
| `retard_jours` | int | Nombre de jours de retard (0 = à temps) |

### 📄 Fichiers 4, 5, 6
- **`services.csv`** — 10 services hospitaliers avec capacité en lits
- **`fournisseurs.csv`** — 5 fournisseurs avec délais et taux de service
- **`ruptures_stock.csv`** — 85 événements de rupture avec impact clinique et patients affectés
